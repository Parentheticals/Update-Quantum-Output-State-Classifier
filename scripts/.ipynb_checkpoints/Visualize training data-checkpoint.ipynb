```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def visualize(data, title, date, color):
    x_val = []
    y_val = []
    
    for d in data:
        # extract real part
        x = [ele.real for ele in d]
        x_val.extend(x)
        # extract imaginary part
        y = [ele.imag for ele in d]
        y_val.extend(y)
        
    plt.scatter(x_val, y_val, 0.1, c = color)
    plt.xlabel('Real Component (Q)')
    plt.ylabel('Imaginary Component (I)')
    plt.ylim([-20, 30])
    plt.xlim([-15, 27])
    plt.title(f'{title} | {date}')
    plt.show()
    
    return (x_val, y_val)
```

```python
date_dir = ['../data2/2020-05-21', '../data2/2020-05-22', '../data2/2020-05-23', '../data3/2021-10-28', '../data3/2021-10-29', '../data3/2021-10-30', '../data3/2021-10-31', '../data3/2021-11-01', '../data3/2021-11-02', '../data3/2021-11-03', '../data3/2021-11-04', '../data3/2021-11-05', '../data3/2021-11-06', '../data3/2021-11-07', '../data3/2021-11-08']
```

```python
## Visualize all the measurements combined in the training data

for date in date_dir:
    train_d_1 = np.load(date + '/train_d_1.npy')
    visualize(train_d_1, "Training Data | I vs Q", date[9:], 'b')
```

```python
## Visualize all the measurements combined in the evaluation data
for date in date_dir:
    valid_d_1 = np.load(date + '/valid_d_1.npy')
    visualize(valid_d_1, "Evaluation Data | I vs Q", date[9:], 'b')
```

```python
## Only want to visualize updated measurements
date = ['../data3/2021-10-29', '../data3/2021-10-30', '../data3/2021-10-31', '../data3/2021-11-01', '../data3/2021-11-02', '../data3/2021-11-03', '../data3/2021-11-04', '../data3/2021-11-05', '../data3/2021-11-06', '../data3/2021-11-07', '../data3/2021-11-08']
state_0 = []
state_1 = []
```

```python
## Visualize measurements in IQ plan for 100% state 0
for dd in date:
    train_d_1_state_0 = np.load(dd + '/train_d_1_state_0.npy')    
    state_0.append(visualize(train_d_1_state_0, 'I vs Q | Prob of state 0 = 100%', dd[9:], 'b'))
```

```python
## Visualize measurements in IQ plan for 100% state 1
for dd in date:
    train_d_1_state_1 = np.load(dd + '/train_d_1_state_1.npy')
    state_1.append(visualize(train_d_1_state_1, 'I vs Q | Prob of state 1 = 100%', dd[9:], 'r'))
```

```python
## Combine the IQ plots of 100% state 0 and 100% state 1
for p in range(len(state_0)):
    plt.scatter(state_0[p][0], state_0[p][1], 0.1, c = 'b')
    plt.scatter(state_1[p][0], state_1[p][1], 0.1, c = 'r')
    plt.ylim([-20, 30])
    plt.xlim([-10, 27])
    plt.title(f'I vs Q. 100% state 0 and state 1 measurements combined | {date[p][9:]}')
    plt.xlabel('Real Component (Q)')
    plt.ylabel('Imaginary Component (I)')
    plt.show()
```
